```python
import matplotlib.pyplot as plt

# Data for the traditional allocation
traditional_allocation = [100, 100, 40]

# Data for the new algorithm allocation
new_algorithm_allocation = [80, 80, 80]

# Bus labels
buses = ['Bus A', 'Bus B', 'Bus C']

# Plotting the data
x = range(len(buses))

plt.figure(figsize=(10, 6))

# Bar plots for traditional and new algorithm allocations
plt.bar(x, traditional_allocation, width=0.4, label='Traditional Allocation', color='lightblue', align='center')
plt.bar([i + 0.4 for i in x], new_algorithm_allocation, width=0.4, label='New Algorithm Allocation', color='lightgreen', align='center')

# Adding labels and title
plt.xlabel('Buses')
plt.ylabel('Occupancy Percentage')
plt.title('Comparison of Bus Occupancy: Traditional vs. New Algorithm')
plt.xticks([i + 0.2 for i in x], buses)
plt.ylim(0, 120)

# Adding the legend
plt.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the traditional allocation
traditional_allocation = [52]*38 + [24] + [0]*11

# Data for the new algorithm allocation
new_algorithm_allocation = [40]*50

# Plotting pie charts
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))

# Traditional Allocation Pie Chart
ax1.pie(traditional_allocation, labels=[f'Bus {i+1}' for i in range(50)], autopct='%1.1f%%', startangle=140)
ax1.set_title('Traditional Allocation')

# New Algorithm Allocation Pie Chart
ax2.pie(new_algorithm_allocation, labels=[f'Bus {i+1}' for i in range(50)], autopct='%1.1f%%', startangle=140)
ax2.set_title('New Algorithm Allocation')

# Display the charts
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the Traditional Allocation (Assuming percentage comparison)
traditional_allocation = [39, 11]  # 39 buses used, 11 buses empty
traditional_fuel = [60, 40]  # 60% more fuel consumption due to inefficiency
traditional_emissions = [70, 30]  # 70% higher carbon emissions

# Data for the Genetic Algorithm Allocation
genetic_allocation = [50, 0]  # All 50 buses used effectively
genetic_fuel = [40, 60]  # 40% fuel consumption due to efficiency
genetic_emissions = [30, 70]  # 30% lower carbon emissions

# Plotting the pie charts
fig, axs = plt.subplots(3, 2, figsize=(14, 18))

# Traditional Allocation Pie Charts
axs[0, 0].pie(traditional_allocation, labels=['Used Buses', 'Empty Buses'], autopct='%1.1f%%', startangle=140)
axs[0, 0].set_title('Traditional Allocation: Bus Utilization')

axs[1, 0].pie(traditional_fuel, labels=['Fuel Used', 'Fuel Saved'], autopct='%1.1f%%', startangle=140)
axs[1, 0].set_title('Traditional Allocation: Fuel Consumption')

axs[2, 0].pie(traditional_emissions, labels=['Carbon Emissions', 'Reduction'], autopct='%1.1f%%', startangle=140)
axs[2, 0].set_title('Traditional Allocation: Carbon Emissions')

# Genetic Algorithm Allocation Pie Charts
axs[0, 1].pie(genetic_allocation, labels=['Used Buses', 'Empty Buses'], autopct='%1.1f%%', startangle=140)
axs[0, 1].set_title('Genetic Algorithm: Bus Utilization')

axs[1, 1].pie(genetic_fuel, labels=['Fuel Used', 'Fuel Saved'], autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Genetic Algorithm: Fuel Consumption')

axs[2, 1].pie(genetic_emissions, labels=['Carbon Emissions', 'Reduction'], autopct='%1.1f%%', startangle=140)
axs[2, 1].set_title('Genetic Algorithm: Carbon Emissions')

# Display the charts
plt.tight_layout()
plt.show()
```
